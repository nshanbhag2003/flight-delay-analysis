# Cleaned Flight Dataset

```python
import pandas as pd

# Load in the clean flight dataset

INPUT_FILE = "/content/drive/My Drive/Classes/Spring 2026/GR5243/Project 1/data cleaning/flights_clean.csv"

df_flights = pd.read_csv(INPUT_FILE)
df_flights.columns = df_flights.columns.str.strip()

print("Rows:", f"{len(df_flights):,}")
print("Columns:", len(df_flights.columns))
print("\nColumn names:")
print(df_flights.columns.tolist())

df_flights.head(10)
```

# Step 1: Time-Based Features

```python
# 1) Convert flight date properly
df_flights['fl_date'] = pd.to_datetime(df_flights['fl_date'])

# 2) Day Name + Weekend Flag
df_flights['day_name'] = df_flights['fl_date'].dt.day_name()

# Derive weekend directly from date
df_flights['is_weekend'] = (
    df_flights['fl_date'].dt.dayofweek >= 5
).astype(int)


# 3) Season feature
def get_season(month):
    if month in [12, 1, 2]:
        return "winter"
    elif month in [3, 4, 5]:
        return "spring"
    elif month in [6, 7, 8]:
        return "summer"
    else:
        return "fall"

df_flights['season'] = df_flights['fl_date'].dt.month.apply(get_season)


# 4) Scheduled departure hour + peak-hour flag

# Convert to datetime
df_flights['crs_dep_time'] = pd.to_datetime(
    df_flights['crs_dep_time'],
    format="%H:%M:%S",
    errors="coerce"
)

# Extract hour cleanly
df_flights['scheduled_dep_hour'] = df_flights['crs_dep_time'].dt.hour

# Peak-hour indicator (morning + evening rush)
df_flights['is_peak_hour'] = (
    df_flights['scheduled_dep_hour'].between(7, 9) |
    df_flights['scheduled_dep_hour'].between(16, 19)
).astype(int)

df_flights.head()
```

# Step 2: Delay Outcome Features

```python
# 1) Delay indicators
df_flights['is_dep_delayed'] = (df_flights['dep_delay'] > 15).astype(int)
df_flights['is_arr_delayed'] = (df_flights['arr_delay'] > 15).astype(int)

# 2) Delay buckets are helpful, so we know if it's minor or major
def delay_bucket(delay):
    if pd.isna(delay):
        return "unknown"
    elif delay <= 15:
        return "on_time"
    elif delay <= 60:
        return "minor"
    elif delay <= 180:
        return "major"
    else:
        return "extreme"

df_flights['delay_bucket'] = df_flights['arr_delay'].apply(delay_bucket)

df_flights.head()
```

# Step 3: Operational Efficiency Features

```python
# 1) Schedule Padding
df_flights['schedule_padding'] = \
    df_flights['crs_elapsed_time'] - df_flights['air_time']

# 2) Excess Elapsed Time
df_flights['excess_elapsed_time'] = \
    df_flights['actual_elapsed_time'] - df_flights['crs_elapsed_time']

# 3) Taxi Metrics
df_flights['taxi_total'] = df_flights['taxi_out'] + df_flights['taxi_in']
df_flights['taxi_share_of_trip'] = \
    df_flights['taxi_total'] / df_flights['actual_elapsed_time']

# 4) Air Speed (some indicator of congestion I suppose)
df_flights['air_speed_mph'] = \
    df_flights['distance'] / (df_flights['air_time'] / 60)

df_flights.head()
```

# Step 4: Route-Level Features

```python
# 1) Route ID
df_flights['route_id'] = \
    df_flights['origin'] + "_" + df_flights['dest']

# 2) Distance Bucket (tells how long of a flight)
def distance_bucket(d):
    if d < 500:
        return "short"
    elif d < 1500:
        return "medium"
    else:
        return "long"
df_flights['route_distance_bucket'] = \
    df_flights['distance'].apply(distance_bucket)

# 3) Route Average Delay
route_delay = df_flights.groupby('route_id')['arr_delay'].mean()
df_flights['route_avg_arr_delay'] = \
    df_flights['route_id'].map(route_delay)

df_flights.head()
```

# Step 5: Airport-Based Features

```python
# 1) Origin Level Delays
origin_delay = df_flights.groupby('origin')['dep_delay'].mean()

df_flights['origin_avg_dep_delay'] = \
    df_flights['origin'].map(origin_delay)

# 2) Delay relative to the airport
df_flights['dep_delay_vs_origin_avg'] = \
    df_flights['dep_delay'] - df_flights['origin_avg_dep_delay']

df_flights.head()
```

# Step 6: Carrier-Level Features

```python
carrier_delay = df_flights.groupby('op_unique_carrier')['arr_delay'].mean()

df_flights['carrier_avg_arr_delay'] = \
    df_flights['op_unique_carrier'].map(carrier_delay)

df_flights.head()
```

# Step 7: Delay Attribution Features

```python
import numpy as np

delay_cols = [
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay'
]

# 1) Total delay attributed to the flight
df_flights['total_reported_delay'] = df_flights[delay_cols].sum(axis=1)

# 2) Flag: does this flight have any attributed delay?
df_flights['has_attributed_delay'] = (df_flights['total_reported_delay'] > 0).astype(int)

# 3) Delay shares: set to 0 when total_reported_delay == 0, otherwise compute share
for col in delay_cols:
    share_col = f"{col}_share"
    df_flights[share_col] = np.where(
        df_flights['total_reported_delay'] > 0,
        df_flights[col] / df_flights['total_reported_delay'],
        0.0
    )

# 4) Primary delay cause: only defined when total_reported_delay > 0
df_flights['primary_delay_cause'] = np.where(
    df_flights['total_reported_delay'] > 0,
    df_flights[delay_cols].idxmax(axis=1),
    np.nan
)

df_flights.head()
```

# Step 8: Outlier Handling (Winsorization)

```python
import numpy as np
import pandas as pd

# Ensure the columns I want to winsorize exist
if 'taxi_total' not in df_flights.columns:
    df_flights['taxi_total'] = df_flights['taxi_out'] + df_flights['taxi_in']

if 'excess_elapsed_time' not in df_flights.columns:
    df_flights['excess_elapsed_time'] = df_flights['actual_elapsed_time'] - df_flights['crs_elapsed_time']

# should winsorize these ones
winsor_cols = [
    'taxi_out',
    'taxi_in',
    'taxi_total',
    'excess_elapsed_time'
]

for col in winsor_cols:
    # origin-specific 99th percentile threshold
    p99_origin = df_flights.groupby('origin')[col].quantile(0.99)

    # fallback threshold in case origin cap is missing
    p99_global = df_flights[col].quantile(0.99)

    # cap threshold per row
    cap = df_flights['origin'].map(p99_origin)

    # fallback to global cap where origin cap is missing
    cap = cap.fillna(p99_global)

    df_flights[f'{col}_cap'] = cap

    # outlier flag: original exceeds threshold
    df_flights[f'{col}_outlier_flag'] = (
        df_flights[col].notna() & (df_flights[col] > df_flights[f'{col}_cap'])
    ).astype(int)

    # winsorized value
    df_flights[f'{col}_winsor'] = np.minimum(df_flights[col], df_flights[f'{col}_cap'])

df_flights.head()
```

```python
summary = []
for col in winsor_cols:
    summary.append({
        "feature": col,
        "outlier_rate": df_flights[f"{col}_outlier_flag"].mean(),
        "mean_raw": df_flights[col].mean(),
        "mean_winsor": df_flights[f"{col}_winsor"].mean(),
    })

pd.DataFrame(summary)
```
